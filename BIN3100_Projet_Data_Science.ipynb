```python
# Importation des bibliothèques nécessaires
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder


data = pd.read_csv("sample_data/Customer_Segmentation.csv")
df = data.copy()

# Étape 1 : Exploration initiale
print("Dimensions du dataset :", df.shape)
print("Aperçu des types de données :\n", df.dtypes.value_counts())
print("Pourcentage de valeurs manquantes :\n", (df.isnull().sum() / df.shape[0]) * 100)

# Étape 2 : Identification des colonnes numériques et catégoriques
num_cols = df.select_dtypes(include=["float64", "int64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns

# Étape 3 : Gestion des valeurs manquantes
# Colonnes numériques : Remplacer les NaN par la médiane
imputer_num = SimpleImputer(strategy="median")
df[num_cols] = imputer_num.fit_transform(df[num_cols])

# Colonnes catégoriques : Remplacer les NaN par la modalité la plus fréquente
imputer_cat = SimpleImputer(strategy="most_frequent")
df[cat_cols] = imputer_cat.fit_transform(df[cat_cols])

# Étape 4 : Encodage des variables catégoriques
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Stocker les encodeurs pour une utilisation future

# Étape 5 : Normalisation des colonnes numériques
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Résultats finaux
print("Aperçu des données après pré-processing :\n", df.head())
print("Vérification des valeurs manquantes :\n", df.isnull().sum().sum())
print("Dimensions finales :", df.shape)

```
