```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Chargement des données
train_data = pd.read_csv("sample_data/Customer_Segmentation.csv")
test_data = pd.read_csv("sample_data/Customer_Segmentation_test.csv")
submission_data = pd.read_csv("sample_data/Customer_Segmentation_prediction_submission.csv")

# Étape 1 : Copie des données pour éviter d'affecter les fichiers originaux
train_df = train_data.copy()
test_df = test_data.copy()

# Étape 2 : Identification des colonnes numériques et catégoriques à partir des données d'entraînement
num_cols = train_df.select_dtypes(include=["float64", "int64"]).columns
cat_cols = train_df.select_dtypes(include=["object"]).columns

# Exclure la colonne Segmentation si present dans cat_cols
cat_cols = cat_cols[cat_cols != 'Segmentation'] # obligé car Segmentation n'est pas present dans test_df

# Étape 3 : Prétraitement sur le dataset de test

# 3.1 Gestion des valeurs manquantes pour les colonnes numériques et catégoriques
imputer_num = SimpleImputer(strategy="median")
imputer_cat = SimpleImputer(strategy="most_frequent")

test_df[num_cols] = imputer_num.fit_transform(test_df[num_cols])
test_df[cat_cols] = imputer_cat.fit_transform(test_df[cat_cols])


# 3.2 Encodage des variables catégoriques
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    le.fit(train_df[col])  # Apprentissage des catégories à partir des données d'entraînement

    # Vérification si les valeurs de test sont bien dans l'entraînement
    test_labels = test_df[col]
    test_df[col] = test_labels.apply(lambda x: x if x in le.classes_ else le.classes_[0])  # Remplacer les valeurs inconnues par le premier label

    test_df[col] = le.transform(test_df[col])
    label_encoders[col] = le


# 3.3 Normalisation des colonnes numériques
scaler = StandardScaler()
test_df[num_cols] = scaler.fit_transform(test_df[num_cols])

# Résultat final pour le dataset de test
print("Dataset de test après pré-processing :")
print(test_df.head())

# Étape 4 : Préparation du fichier de soumission des prédictions
submission_data["Segmentation"] = submission_data["Segmentation"].apply(
    lambda x: x if x in label_encoders[cat_cols[0]].classes_ else label_encoders[cat_cols[0]].classes_[0]
)
submission_data["Segmentation"] = label_encoders[cat_cols[0]].transform(submission_data["Segmentation"])

# Résultats finaux pour train_data
print("Aperçu des données après pré-processing :\n", train_df.head())
print("Vérification des valeurs manquantes :\n", train_df.isnull().sum().sum())
print("Dimensions finales :", train_df.shape)


## Résultats finaux pour test_data
print("Dataset de test après pré-processing :\n", test_df.head())
print("Vérification des valeurs manquantes :\n", test_df.isnull().sum().sum())
print("Dimensions finales :", test_df.shape)

```
